```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
from langchain.chains.question_answering import load_qa_chain
from langchain.docstore.document import Document

# Load a pre-trained QA model from Hugging Face
qa_model = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Wrap the Hugging Face pipeline in a LangChain LLM class
llm = HuggingFacePipeline(pipeline=qa_model)

# Sample text
text = """
LangChain is a framework for developing applications powered by language models.
It enables developers to easily integrate these models with various data sources and workflows.
"""

# Ask a question
question = "What is LangChain?"

# Format the input as a dictionary
input_dict = {"question": question, "context": text}

# Get the answer directly from the pipeline (not using a chain)
# This is because the Hugging Face pipeline expects a specific input format
answer = qa_model(input_dict)

print("Question:", question)
print("Answer:", answer['answer'])
```

```python
# Define a knowledge base with keywords and their synonyms
knowledge_base = {
    "langchain": "LangChain is a framework for developing applications powered by language models.",
    "python": "Python is a versatile programming language.",
    "data science": "Data science is an interdisciplinary field that uses scientific methods."
}

# Define synonyms
synonyms = {
    "langchain": ["lang chain", "language chain"],
    "python": ["py", "pythonic"],
    "data science": ["data-science", "ds"]
}

def simple_qa_system(question):
    question = question.lower()

    for keyword in knowledge_base:
        # Check if the keyword or any of its synonyms appear in the question
        if keyword in question or any(synonym in question for synonym in synonyms.get(keyword, [])):
            return knowledge_base[keyword]

    return "Sorry, I don't know the answer to that question."

# Ask a question
question = "What is Pythonic?"

# Get the answer
answer = simple_qa_system(question)

print("Question:", question)
print("Answer:", answer)
```
