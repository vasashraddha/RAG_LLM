```python
import pandas as pd
import chromadb
from sentence_transformers import SentenceTransformer
import numpy as np
import os
```

```python
df=pd.read_csv('/content/the_oscar_award.csv')
df
```

```python
df=df.loc[df['year_ceremony'] == 2024]
df
```

```python
df=df.dropna(subset=['film'])
df.loc[:, 'category'] = df['category'].str.lower()
df
```

```python
df.loc[:, 'text'] = df['name'] + ' got nominated under the category, ' + df['category'] + ', for the film ' + df['film'] + ' to win the award'
df.loc[df['winner'] == False, 'text'] = df['name'] + ' got nominated under the category, ' + df['category'] + ', for the film ' + df['film'] + ' but did not win'
df
```

```python
client = chromadb.Client()
collection = client.get_or_create_collection("oscars-2024")
```

```python
docs=df["text"].tolist()
```

```python
ids= [str(x) for x in df.index.tolist()]
```

```python
collection.add(
    documents=docs,
    ids=ids
)
```

```python
results = collection.query(
    query_texts=["Poor Things"],
    n_results=1
)
```
